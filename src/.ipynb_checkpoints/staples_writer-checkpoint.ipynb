```python
import pandas as pd
```

```python
df2 = pd.read_csv('staples_master_data.csv', index_col=0)
df2
```

```python
p_1 = list(df2.loc[df2['Priority'] == 1]['Item'])
p_1
```

```python
p_2 = list(df2.loc[df2['Priority'] == 2]['Item'])
p_2
```

```python
p_3 = list(df2.loc[df2['Priority'] == 3]['Item'])
p_3
```

```python
p_4 = list(df2.loc[df2['Priority'] == 4]['Item'])
p_4
```

```python
p_5 = list(df2.loc[df2['Priority'] == 5]['Item'])
p_5
```

```python
p1 = [p.strip() for p in p_1]
p1
```

```python
p2 = [p.strip() for p in p_2]
p2
```

```python
p3 = [p.strip() for p in p_3]
p3
```

```python
p4 = [p.strip() for p in p_4]
p4
```

```python
p5 = [p.strip() for p in p_5]
p5
```

```python
with open('staples_list.txt', 'w+') as out:
    out.write(','.join(p1))
    out.write('\n')
    out.write(','.join(p2))
    out.write('\n')
    out.write(','.join(p3))
    out.write('\n')
    out.write(','.join(p4))
    out.write('\n')
    out.write(','.join(p5))
    out.write('\n')
```

```python
with open('staples_list.txt', mode = 'r', encoding='utf-8') as f:
    staples = f.readlines()
    
staples
```

```python
for key, value in enumerate(staples):
    print(key, value)
```
