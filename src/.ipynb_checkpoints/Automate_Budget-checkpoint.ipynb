```python
df['Item'] = df['Item'].str.strip()
```

```python
df_bk = df.copy()
```

```python
for key, value in enumerate(staples):
    
    lst = value.split('\n')
    lst = lst[:-1]
    lst = value.split(',')
    lst[-1] = lst[-1].strip()
    l = df[df[['Item']].isin(lst).any(axis=1)].copy()
    l['Priority'] = key + 1
    df.update(l)
```

# Randomize

> We are only taking items of Priority 1, 2 and 3. Priority Item 4 is more discretionary and Priority Item 5 is very very rarely purchased. 

```python
budget = pd.DataFrame()
```

```python
df['Price'] = pd.to_numeric(df['Price'])
```

```python
def budget_calculate():
    total_so_far = budget['Price'].sum()
    return total_so_far
```

## Staples are Priority 1 items - all are added

```python
df_p = df.loc[df['Priority'] == 1].copy()
df_p
```

```python
budget = df_p.copy()
budget
```

```python
amt_spent = budget_calculate()
budget_remaining = budget_amt - amt_spent
print('Budget Remaining = ', budget_remaining, ', Amount Spent =', amt_spent)
```

## Priority 2 - One specific category - we take 2 items 

```python
df_p = df.loc[df['Priority'] == 2].copy()
df_p
```

```python
df_specific = df_p[df_p['Category'] == 'regular vegetables']
df_specific
```

```python
df_specific = df_specific.sample(2)
df_specific
```

```python
budget = budget.append(df_specific)
budget
```

```python
amt_spent = budget_calculate()
budget_remaining = budget_amt - amt_spent
print('Budget Remaining = ', budget_remaining, ', Amount Spent =', amt_spent)
```

## Priority 2 items - All categories

> With randomized sampling there is a possibility that a record chosen in this section can be a duplicate of one of the items chosen in the section before.

```python
df_specific = df_p.groupby(['Category'])['Item', 'UnitAmt', 'Unit', 'Price', 'Priority'].apply(pd.Series.sample, n=1).reset_index(level=[0])
df_specific
```

```python
budget = budget.append(df_specific)
budget
```

```python
amt_spent = budget_calculate()
budget_remaining = budget_amt - amt_spent
print('Budget Remaining = ', budget_remaining, ', Amount Spent =', amt_spent)
```

## Priority 3 items 

```python
df_p = df.loc[df['Priority'] == 3].copy()
df_p
```

```python
df_specific = df_p.groupby(['Category'])['Item', 'UnitAmt', 'Unit', 'Price', 'Priority'].apply(pd.Series.sample, n=1).reset_index(level=[0])
df_specific
```

> With Priority 3 items - we add each item one by one until we match the budget or just cross it. Even with replace=false just like it Priority Item 2, there is a possibility of duplicates

```python
while amt_spent < budget_amt:
    sample = df_specific.sample(1, replace=False)
    budget = budget.append(sample)
    amt_spent = budget_calculate()
    budget_remaining = budget_amt - amt_spent
    print('Budget Remaining = ', budget_remaining, ', Amount Spent =', amt_spent)
```

```python
budget
```

# Print finalized order

```python
# TODO - Can add print sequence if required
budget = budget.sort_values(by=['Category','Priority'])
budget['UnitAmt'] = budget['UnitAmt'].astype('int')
category = ' '

with open('weekly_order.txt', 'w') as out:
    for row in budget.itertuples():
        if category != row.Category:
            category = row.Category
            header = '\n**' + row.Category + '**'
            print(header)
            print(header, file = out)
        line = str(row.Item) + ' - ' + str(row.UnitAmt) + ' ' + str(row.Unit) 
        print(line)
        print(line, file = out)
    total_line = '\nTOTAL = ' + str(budget['Price'].sum())
    print(total_line)
    print(total_line, file=out)
```

## Connect to database

```python
import sqlite3
from sqlite3 import Error
```

## Write the master and order data to database

```python
database = 'weekly_budget.db'
from contextlib import closing

with closing(sqlite3.connect(database)) as connection:
    df.to_sql('master', con = connection, if_exists='append', index=False)
    budget.to_sql('weekly_order', con = connection, if_exists='append', index=False)
```

```python
def create_connection(db_file):
    """ create a database connection to a SQLite database """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
        return conn
    except Error as e:
        print(e)
```

```python
conn = create_connection(database)
```

```python
check_master = pd.read_sql("select * from master", con=conn)
print(check_master)
```

```python
check_weekly = pd.read_sql("select * from weekly_order", con=conn)
print(check_weekly)
```
